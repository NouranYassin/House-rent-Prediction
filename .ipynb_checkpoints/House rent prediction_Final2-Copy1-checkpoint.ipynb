#### Split data to trian & Test

```python
# Drop UnNecessary variables:
df.drop( columns = ['Posted_On' ,'Floor', 'Area_Locality', 'year_posted_on'] , axis = 1 , inplace = True)
```

```python
df.shape
```

```python
## Convert them to string to work as a categorical features
df['BHK'] = df['BHK'].astype(str)
df['day_posted_on'] = df['day_posted_on'].astype(str)
df['month_Posted_on'] = df['month_Posted_on'].astype(str)
df['Bathroom'] = df['Bathroom'].astype(str)
```

```python
# Split to features and target 
X = df.drop(columns = 'Rent' , axis = 1)
y = df['Rent']
```

```python
df.shape
```

```python
# Split to train and test 
X_train , X_test , y_train , y_test =  train_test_split(X , y , test_size = 0.2 , shuffle = True , random_state = 45 )

print('X_train.shape ' , X_train.shape)
print('y_train.shape ' , y_train.shape)
print('X_test.shape ' , X_test.shape)
print('y_test.shape ' , y_test.shape)
```

```python
X_train.info()
```

```python
X_train
```

```python
##slice the lists
num_cols = X_train.select_dtypes(include='number').columns.tolist()

categ_cols_nominal = ['Area_Type','City','Tenant_Preferred','Point_of_Contact','BHK','Bathroom','month_Posted_on', 'day_posted_on']
categ_cols_ordinal = ['Furnishing_Status','size_category']

 
```

```python
## My Pipeline:


## Numerical -------------> Impute (median), standrize (RobustScaler)
## Categorical: Nominal --> Impute (mode), Encoding Label Encoding
## Categorical: Ordinal --> Impute (mode), Encoding Ordinal Encoding

## For Numerical
num_pipeline = Pipeline(steps=[
                                ('selector', DataFrameSelector(num_cols)),
                                ('Imputer', SimpleImputer(strategy='median')),
                                ('standrdize',RobustScaler())
                            ])


## For Categorical: Nominal
categ_nominal_pipeline = Pipeline(steps=[
                                ('selector', DataFrameSelector(categ_cols_nominal)),
                                ('imputer', SimpleImputer(strategy='most_frequent')),
                                ('Encoding', OrdinalEncoder())
                                        ])

## For Categorical: Ordinal
categ_ordinal_pipeline1 = Pipeline(steps=[
                                ('selector', DataFrameSelector(categ_cols_ordinal)),
                                ('imputer', SimpleImputer(strategy='most_frequent')),
                                ('encoding',OrdinalEncoder())
                                        ])


## Combine all
all_pipeline = FeatureUnion(transformer_list=[
                        ('numerical', num_pipeline),
                        ('categ_nominal', categ_nominal_pipeline),
                        ('categ_ordinal1', categ_ordinal_pipeline1),


                        ])

## apply
X_train_final = all_pipeline.fit_transform(X_train)
X_test_final = all_pipeline.transform(X_test)
```

```python
X_train_final.shape, X_test_final.shape
```

#### `Building Models`
`Let's make our evaluation metric to be MSE`

### Linear Regression

```python
## Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_final, y_train)


## predict
y_pred_lin_train = lin_reg.predict(X_train_final)
y_pred_lin_test = lin_reg.predict(X_test_final)


## Evaluate
RMse_lin_train = np.sqrt(mean_squared_error(y_train, y_pred_lin_train))
RMse_lin_test = np.sqrt(mean_squared_error(y_test, y_pred_lin_test))

print(f'RMean Sqaure error using LinearRegression for Training is {RMse_lin_train  :.3f}')
print(f'RMean Sqaure error using LinearRegression for Test is {RMse_lin_test :.3f}')
```

Model is Overfit so we reduce complexity of the upcoming models

### Polynomial Regression

```python
n = 11
d = 2

import math

math.factorial(n+d) / (math.factorial(d) * math.factorial(n))
```

```python
## prprocessing
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train_final)
X_test_poly = poly.transform(X_test_final)

## Linear Regression
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_train_poly, y_train)

## predict
y_pred_poly_train = lin_reg_poly.predict(X_train_poly)
y_pred_poly_test = lin_reg_poly.predict(X_test_poly)

## Evaluate
RMse_poly_train = np.sqrt(mean_squared_error(y_train, y_pred_poly_train))
RMse_poly_test = np.sqrt(mean_squared_error(y_test, y_pred_poly_test))


print(f'RMSE using Polynomial Regression for Training is {RMse_poly_train :.3f}')
print(f'RMSE using Polynomial Regression for Test is {RMse_poly_test  :.3f}')
```
